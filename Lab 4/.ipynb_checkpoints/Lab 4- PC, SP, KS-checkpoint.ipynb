## Assignment 1 --  Matched FIlter with Ideal Signals (Title of Assignment 1)

```python
# Assignment 1 - Time Scaling Function

fs=8000 # sampling frequency

# epsilon must always be less then 1/fs!! Otherwise, the time vector will not be in range!!
epsilon=1/(fs + 1)

def u(t):
    return 1.0*(t>=0)

t = np.arange(0, 1 + epsilon, 1/fs)
# 1.
u_t = u(t)
u_t1 = u(t-1)
s1 = u_t - u_t1
s0 = -s1

# 2.
y1 = np.convolve(s1, s1)/fs
y0 = np.convolve(s1, s0)/fs

# 3.
t_y = np.arange(0, len(y1)/fs, 1/fs)
fig = plt.figure(1)
plt.subplot(2, 1, 1)
plt.xlabel('Time(s)')
plt.ylabel('y1')
plt.title('y1 vs. Time')
plt.plot(t_y, y1)

plt.subplot(2, 1, 2)
plt.xlabel('Time(s)')
plt.ylabel('y0')
plt.title('y0 vs. Time')
plt.plot(t_y, y0)

plt.tight_layout()

# 4.
y1_corr = y1[len(s1) - 1]
y0_corr = y0[len(s0) - 1]

# Discussion
```

## Assignment 2 -- Signal Decoding with a Matched Filter (title of assignment 2)

```python
# A.
def decode(r, fs, s0, s1):
    y0 = np.convolve(r, s0)/fs
    y1 = np.convolve(r, s1)/fs
    y0_corr = y0[len(s0) - 1]
    y1_corr = y1[len(s1) - 1]
    s = 0
    if (y1_corr > y0_corr):
        s = 1
    return s
    
```

```python
# B.a
n = np.random.normal(0,10,len(s1))

# B.b
r = s1 + n
t_r = np.arange(0, len(r)/fs, 1/fs)

# B.c
fig = plt.figure(2)
plt.subplot(2, 1, 1)
plt.title('r(t) vs t')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.xlim(0,2)
plt.plot(t_r, r)
```

```python
# C.a
yr1 = np.convolve(r, s1) / fs
yr0 = np.convolve(r, s0) / fs

t_yr1 = np.arange(0, len(yr1)/fs, 1/fs)
t_yr0 = np.arange(0, len(yr0)/fs, 1/fs)

# C.b
plt.subplot(2, 1, 2)
plt.title('Convolution with Noisy Signal')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.plot(t_yr1, yr1, label="r(t) * s1(t)", color="magenta")
plt.plot(t_yr0, yr0, label="r(t) * s0(t)", color="red")
plt.legend(loc="upper left")
plt.xlim(0,2)
plt.tight_layout()

# C.c
yr1_corr = yr1[len(s1)-1]
yr0_corr = yr0[len(s0)-1]

print("yr1_corr:" + str(yr1_corr))
print("yr0_corr:" + str(yr0_corr))
```

```python
#D
print(decode(s1+n, fs, s0, s1))
print(decode(s0+n, fs, s0, s1))
```

###  Discussion

The correlation measurement between s1(t) and s0(t) would be less than the correlation measurement when both signals were box signals. When s0(t) is not a box signal, the max correlation values are 0.5 and -0.5 rather than 1 and -1. When s0(t) is a box signal, it is more useful for communication because the correlation has one distinct peak of -1, and its peak is at the midpoint.
